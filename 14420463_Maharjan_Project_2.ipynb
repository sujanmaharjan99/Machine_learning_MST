```python
def count_words_in_file(file_path, words_to_ignore):
    with open(file_path, 'r', errors='ignore') as file:
        text = file.read().lower()
        words_in_email = text.split()
        word_count = {}
        unique_word_count=0
        for word in words_in_email:
            if word not in words_to_ignore:
                unique_word_count+=1
                word_count[word] = word_count.get(word, 0) + 1
        word_count['<TOTAL_UNIQUE_WORDS>'] = unique_word_count
        word_count['<TOTAL_WORDS>'] = len(words_in_email)
        #if there is already existing class then add the word counts to the class
        total_words_in_class = sum(word_count.values())

    return word_count

def likelihood_estimation(word_count, total_words_in_class, vocab_size, alpha=1):
    likelihoods = {}
    for word, count in word_count.items():
        likelihoods[word] = (count + alpha) / (total_words_in_class + alpha * vocab_size)
    return likelihoods
```
